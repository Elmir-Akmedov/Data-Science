```python
df.drop("emp no", inplace=True, axis=1)
df.drop("Employee Number", inplace=True, axis=1)
df.drop("Over18", inplace=True, axis=1)
df.drop("Standard Hours", inplace=True, axis=1)
df.drop("Employee Count", inplace=True, axis=1)
df.drop("CF_attrition label", inplace=True, axis=1)
df.drop("CF_age band", inplace=True, axis=1)
df.drop("CF_current Employee", inplace=True, axis=1)
df.drop("Department", inplace=True, axis=1)
```

```python
df["avg_salary_for_job"] = df.groupby("Job Role")["Monthly Income"].transform("mean")
df["ratio_of_salary"] = df["Monthly Income"] / df["avg_salary_for_job"]
df["avg_salary_for_job_level"] = df.groupby("Job Level")["Monthly Income"].transform("mean")
df["ratio_of_salary_by_level"] = df["Monthly Income"] / df["avg_salary_for_job_level"]
```
